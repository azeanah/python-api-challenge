```python
# Dependencies and setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import json
import os

# Import API key
from api_keys import g_key

# configure gmaps
gmaps.configure(api_key=g_key)

#import citipy
import citipy
```

```python
# load csv in P1 into a DF
cities_pd = pd.read_csv("output_data/cities_aaron.csv")
cities_pd.dropna(inplace = True) 
cities_pd.head()
```

```python
locations = cities_pd[["Lat", "Lon"]]
humidity = cities_pd["Humidity"].astype(float)
```

```python
# heat map layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=5, point_radius=1)

figure_layout = {
    'width': '400px',
    'height': '300px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
#define ideal parameters

high_temp = cities_pd.loc[cities_pd["Temp"] > 80].index
low_temp = cities_pd.loc[cities_pd["Temp"] < 70].index
mid_temp = cities_pd.drop(high_temp)

ideal_temp = mid_temp.drop(low_temp)
ideal_wind = cities_pd.loc[cities_pd["Wind Speed"] < 10]
ideal_cloud = cities_pd.loc[cities_pd["Cloudiness"] == 0]

```

```python
# create new dataframe fitting weather criteria
# narrow down the cities to fit weather conditions
# drop any rows with null values

id_temp_wind = ideal_temp.loc[ideal_temp["Wind Speed"] < 10]

ideal_df = id_temp_wind.loc[id_temp_wind["Cloudiness"] == 0]

ideal_df
```

```python
#target_coordinates = ideal_df[["Lat", "Lon"]]
target_radius = 5000
target_type = "lodging"
target_search = "hotel"

params = {
#    "location": target_coordinates,
    "keyword": target_search,
    "radius": target_radius,
    "type": target_type,
    "key": g_key
}

g_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

#response = requests.get(g_url, params=params)

#places_data = response.json()

#hotel_df = pd.DataFrame["Hotel Name"]
ideal_df["Hotel Name"] = ""
ideal_df["Hotel Location"] = ""
ideal_df["Hotel Lat"] = ""
ideal_df["Hotel Lon"] = ""


hotel_lat = []
hotel_lon = []
#hotel_df
hotel_name = []
#city = []



for index, city in ideal_df.iterrows():
    g_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"
    try:
        lat = city["Lat"]
        lon = city["Lon"]
        params['location'] = f"{lat},{lon}"
        response = requests.get(g_url, params=params)
        places_data = response.json()
        ideal_df.loc[index, "Hotel Name"] = places_data["results"][0]["name"]
        #ideal_df.loc[index, "Lon"] = places_data[["results"][0]["geometry"]["location"]["lng"]
        #hotel_name.append(response['name'])
        ideal_df.loc[index, "Hotel Lat"] = places_data["results"][0]["geometry"]["location"]["lat"]
        ideal_df.loc[index, "Hotel Lon"] = places_data["results"][0]["geometry"]["location"]["lng"]
        print("Hotel Name")
    except (KeyError, IndexError):
        print("Not found...skipping.")

    print("------------")

hotel_location = str(hotel_lat) + str(hotel_lon)
```
